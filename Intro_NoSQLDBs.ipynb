##**En qué consite NoSQL**

El término NoSQL fue introducido en un evento para discutir todas las bases de datos open-sources que estaban apareciendo en escena, y el nombre de NoSQL se le acuño desde entonces.

Por desgracia, el término NoSQL hace referencia a lo que no significa de forma literal. Se refiere más bien, a una familia de bases de datos que varian apliamente en estilo y tecnología, pero todas comparte en común que no son de naturaleza relacional, lo que significa que no son un RDBMS (Relational Database Managment System) estandar de filas y columnas. En este sentido, un nombre que mejor describiría a este tipo de bases de datos es que son **no relacionales**.


NoSQL $\rightarrow$ "Not Only SQL" (DB no relacionales)

Las bases de datos NoSQL proporcionan nuevas formas de guardar y consultar datos que aborda una serie de problemas para las aplicaciones modernas. Lo más importante es que la mayoria de bases de datos NoSQL están orientadas a manipular una variedad diferente de problemas de escala que han surgido asociados al tratamiento de grandes datos (big data). Con la *escala* nos referimos tanto al tamaño como la manipulación de varios usuarios en los datos.

##**Breve historia de las bases de datos**

Dentro del desarrollo de software, las bases de datos surgen de la neesidad de guardar datos relacionados con una aplicación. Hace 10 años o más, los desarrolladores escogía entre una serie de bases de datos relacionales como:

- Oracle
- MySQL
- SQL Server
- DB2

Posteriormente, entre los 90 y principios del 2000, se da el boom de *dotcom*, por lo que las aplicaciones de internet y las compañias pasan de servir un puñado de empleados en las empresas a millones de usuarios en el internet público. Para estas aplicaciones, la **disponibilidad** y el **rendimiento** fueron primordiales y estos nuevos problemas de escala llevaron a un impulso para crear nuevas tecnologías escalables para dichos datos. Algunas de las tecnologías que aparecen en ese entonces son:

- MapReduce de Google, para el procesamiento de grandes datos en sistemas distribuidos.
- Dynamo de Amazon, para la distribución de datos y gestión de carga de trabajos en un cluster de un estilo de arquitectura quorum.

A finales del 2000, surgen varias bases de datos omo

- Apache CouchDB
- Cassandra
- Hbases
- Mongo
- Riak

las cuales se vuelven particularmente útilies para el uso de aplicaciones que requieren datos a gran escala, y que son más faciles de manejar que las bases de datos relacionales. 

En los últimos 5 años, varias bases de datos NoSQL han aprovechado un modelo de servicio completamente administrado, también llamado base de datos como servicio, para reducir la administración y el mantenimiento del usuario final y permitir a los desarrolladores centrarse en crear aplicaciones con estas bases de datos modernas.

###**Key-Value**

####**Arquitectura**

Las bases de datos **Key-Value** son las menos complejas de todas de los tipos de bases de datos NoSQL. Todos los datos se almacenan con una clave (key) y un blob de valor asociado.

**Pros:**

Debido a que el almacenamiento de forma Key-Value representa un hashmap (array a cuyos elementos se puede acceder mediante un número o un string), son potentes para realizar operaciones básicas de CRUD (Create-Read-Update-Delete), y, generalmente, se escalan bastante bien y se fragmentan fácilmente en un número $x$ de nodos.

**Contras:**

Cada **fragmento** contendrá un rango dado de claves y valores asociados. Sin embargo, estas bases de datos no se usan para consultas complejas que intentan conectar varias piezas de datos, y son atómicos solo para una sola clave. El valor blob es opaco y, por lo tanto, generalmente tendrá menos flexibilidad cuando se trata de indexar y consultar los datos que otros tipos de bases de datos.

**Resumen**:

- Almacenamiento de datos basados en una clave única en un **hashmap**.
- El Valor "blob" asociado con una clave es "opaco" para la base de datos.
- Alto rendimiento en operaciones básicas CRUD, pero rendimiento deficiente en consultas complejas o datos interconectados.
- Característica atómica solo para operaciones de una clave única.
- Escalable horizontalmente a través del fragmentado.

####**Casos de uso**

Cada véz que se necesitemos un buen rendimento para operaciones básicas CRUD y los datos no estén interconectados, la mejor opción de base de datos NoSQL son las de tipo Key-Value.

Ejemplo:

- Almacenamiento y recuperación de información de una aplicación web: Cada sesión de usuario recibirá algún tipo de valor único tipo key y todos los datos serán almacenados juntos en un valor *blob* opaco. Además, no habría necesidad de consultar en función de la información en el blobl de valores. Todas las transacciones se basarían en el key único.
- Un uso similar se prodría presentar en el caso del almacenamiento de perfiles y preferencias de usuarios dentro de una aplicación.
- Almacenamiento de datos en un carrito de compra en línea en las tiendas o mercados.

En estos ejemplos, las consultas o manipulaciones de realaciones complejas entre diferentes claves serían muy pocas y distantes entre sí. Cualquier caso que se aleje de este escenario es inadecuado para la implementación de bases de datos NoSQL basadas en claves.

Los casos ene los que no se debe usar bases de datos NoSQL se resumen a continuación:

- Relaciónes entre los datos de muchos contra muchos,
- Requerimiento de una consistencia de alto nivel para transacciones de múltiples operaciones.
- Si pretende realizar una consulta del valor vs la clave.

### **Document DB**

#### **Arquitectura**

Las bases bases de datos **Document** se basan en el modelo *key-value* al hacer que el valor sea visible y pueda consultarse. 

Cada dato se considera un documento y almacenado como alguno de los siguientes archivos:

- JSON
- XML
- BSON

Una de las ventajas de las bases de datos de **documentos** es que cada **documento** ofrece un esquema flexible, en el que no es necesario que dos documentos sean iguales o contengan la misma información. Las bases de datos de documentos generalmente ofrecen la capacidad de indexar y consultar el contenido de los documentos, ofreciendo búsquedas de rango de valores y claves y capacidad de búsqueda, o quizás consultas analíticas a través de paradigmas como MapReduce.

Las bases de datos de **documentos** son escalables horizontalmente y permiten la fragmentación a través de múltiples nodos, típicamente fragmentados por alguna clave única en el documento. Generalmente, el almacenamiento de documentos solo garantizan transacciones atómicas en operaciones de un solo documento.

**Resumen Document DB**

- Los datos son guardados como documentos: JSON, XML, BSON, etc.
- Los documento son autodescriptivos y los datos tienen una jerarquía de árbol.
- Son similares a las bases de datos Key-Value pero el <font color='red'>valor es visible y examinable</font>.
- Uso de MapReduce para indexado y analítica.
- Esquemas de entrada flexibles ("schema-less")
- Atomicos solo para operaciones de un solo docimento.
- Escalables horisontalmente a través de particiones.




#### **Casos de uso:**

Algunos casos de usos de bases de datos NoSQL  de tipo **Document** son:

- **Registro de eventos**, para una aplicación o proceso: Cada instancia puede constituir un nuevo documento o agregado que contiene toda la información del evento correspondiente.
- **Blogs online y otras aplicaciones web:** Cada usuario se representaría por un documento; cada post como documento; cada comentarior, like, o acción sería un documento. 
- **Datos operacionales y metadatos para aplicaciones web/movil**

Todos los documentos contendrían información relacionada con el tipo de datos, como nombre de usuario, contenido de publicación o timestamp de cuándo se creó el documento. En términos más generales, el almacenamiento de documentos funciona bien con conjuntos de datos operacionales para aplicaciones web y móviles. Estas bases de datos fueron diseñados teniendo en cuenta el internet, pensando en documentos JSON, API RESTful y datos no estructurados.

#### **Caso en los que falla**

Otros modelos de bases de datos pueden ser mejores soluciones cuando se necesitan transacciones ACID. No es posible en el almacenamiento de un documento manejar una transacción que opera en varios documentos y una base de datos relacional puede ser una mejor opción.

En segundo lugar, las bases de datos de cocumentos pueden resultar en una opción incorrexta si se encuentra forzando los datos en un diseño <font color='red'>aggregate-oriented</font>. Si pertenece de forma natural a un modelo normalizado/taular, esto implica la investigación de bases de datos relacionales.

##**Archivos `BSON`**

Un objeto BSON es un formato usado para el intercambio de datos y almacenamiento en la base de datos `MongoDB`. Es una representación binaria de estructuras de datos y mapas, de ahí su nombre BSON que significa *Binary JSON*.

Un objeto *BSON* es una lista ordenada de elementos de diferentes tipos como:

- `string`
- `int64/32`
- Datos reales de 64 bits
- `Date`
- `array` de bytes
- `bool`
- `Null`
- Objeto BSON
- Arreglo BSON
- Expresiones regulares
- Código JavaScript

Los tipos soportados por un objeto `BSON` son los mismos que el caso de un objeto `JSON`, añadiendse los tipos `Date` y arreglos de bytes.

##**Creación de una colección**

Para crear nuestra base de datos nueva, podemos acceder a una base de datos que definamos por defecto, y mediante la el cliente de conexión, crear la base de datos con un nombre nuevo:

````
# Creación colección dentro de una base de datos

from pymongo import MongoClient
from getpass import getpass

def crear_coleccion(db_name = "prueba", new_coll = "toDel"):

  clave = getpass("Digite su clave: ")

  url_db = "mongodb+srv://ssanchez2:{}@cluster0.a463h.mongodb.net/{}?retryWrites=true&w=majority".format(clave, db_name)         
  
  # Conexión
  cliente = MongoClient(url_db)

  #"Cursor"
  db = cliente[db_name]

  db.create_collection(new_coll) 

  # Desconexión

  cliente.close()


if __name__ == "__main__":
  crear_coleccion()

````

````
from pymongo import MongoClient
from tabulate import tabulate
from getpass import getpass

def select_sex(db_name = 'test', coll_name = 'estudiantes', Sex = 'F'):
    
    clave = getpass()
    #Establecemos conexión con la base de datos
    cliente = MongoClient("mongodb+srv://ssanchez2:{}@cluster0.a463h.mongodb.net/{}?retryWrites=true&w=majority".format(clave, db_name))
    
    # Escogemos base de datos "cursor"
    db = cliente[db_name]

    #Buscar todos los documentos
    est_doc = db[coll_name].find({'Sexo': Sex}) 

    #Hacer una cuenta de entradas en la colección
    est_counts = db[coll_name].find({'Sexo': Sex}).count()

    if not est_doc:
        print("No se encontró un estudiantes con esas características" )
    else:
        # Si solo busco uno, o hay uno, descomentar líneas.
        #print(est_doc)
        print("Existen %d documentos en la colección" % est_counts)
        print(tabulate(est_doc, headers = "keys"))
        

    cliente.close() 

if __name__ == "__main__":
    select_sex(db_name='prueba')
````

####**Recuperaciones Simples:**

#####**Coincidencia en un campo**

Con la clausula `where` podemos, por ejemplo, hacer una consulta simple de un registro de los empleados con cierto nombre:

````
#recuperación de documentos

docs = db.collection(f'{coll_name}').where('NOMBRE','==','Ana').stream()


for doc in docs:
  print('{} => {}'.format(doc.id, doc.to_dict()))
````

Si queremos realizar una consulta más compleja, como extraer un empleado que tiene cierto `NOMBRE` y `APELLIDO`, debemos recurrir a las consultas compuestas, como veremos más adelante.

#####**Coincidencia en varios campos**

Si se quiere extraer del ejemplo anterior solo la persona con un apellido dato, realizamos dos veces la operación where:

````
ref_e = db.collection(f'{coll_name}')
docs = ref_e.where('NOMBRE', '==','Ana').where('APELLIDO', '==','Quiroga').stream()

for doc in docs:
  print('{} => {}'.format(doc.id, doc.to_dict()))
````

#####**Organizar por un campo**

Mostremos la lista de empleados organizada por salario

````
# Mostrar lista por salario

ref_e = db.collection(f'{coll_name}')
docs = ref_e.order_by('SALARIO').stream()

for doc in docs:
  print('{} => {}'.format(doc.id, doc.to_dict()))
````

Por defecto la lista se encuentra organizada de forma ascendente. Si queremos organizarlos de forma descendente la búsqueda, lo indicamos a través del argumento basado en la clave `direction` 

````
# Salario de forma Descendente
ref_empleados = db.collection(f'{coll_name}')
docs = ref_empleados.order_by('SALARIO', direction = firestore.Query.DESCENDING).stream()

for doc in docs:
  print('{} => {}'.format(doc.id, doc.to_dict()))
````

#####**Combinación de filtros `where`, `orderBy` y `limit()`**

Podemos complicar aún más las búsquedas. Por ejemplo, podemos hallar las personas que ganen más de cierto salario, organizarlo por salario y limitar la búsqueda a 3 personas:

````
#combinación filtros
referencia_empleados = db.collection(f'{coll_name}')
docs = referencia_empleados.where('SALARIO','>',3000000).order_by('SALARIO').limit(3).stream()

for doc in docs:
  print('{} => {}'.format(doc.id, doc.to_dict()))
````

**Regla importante:**

Si tienes un filtro con una comparación de rangos (`<`, `<=`, `>`o `>=`), el primer ordenamiento que se hace debe ser respecto al mismo campo de la búsqueda, como en el ejemplo anterior.

[Ver link](https://firebase.google.com/docs/firestore/manage-data/transactions)

Una transacción es un conjunto de operaciones de lectura y escritura realizadas en un o más documentos. La transacción se compone por operaciones `get()` seguidas de cualquier número de operaciones de escritura como `set()`, o `delete()`

````
doc = "EmpleadoF"

transaction = db.transaction()
ref_emp = db.collection('Empleados').document(f"{doc}")

@firestore.transactional
def update_in_transaction(transaction, ref_emp):
    snapshot = ref_emp.get(transaction=transaction)
    transaction.update(ref_emp, {
        'EDAD': snapshot.get('EDAD') + 1
    })

update_in_transaction(transaction, ref_emp)
````

Cuando se usen transacciones, se debe tener en cuenta lo siguiente:

- La operación de lectura se debe ejecutar antes que la de escritura.

- Una función que llama a una transacción (fnción de transacción) se podría ejecutar mas de una vz si una edición simultánea afecta a un documento eque la transacción lee.

- Las funciones de transacción no deberían modificar el estado de la app directamente.

- Las transacciones fallarán cual el cliente no tenga conexión.

####**Pasar información fuera de una transacción**

En el siguiente ejemplo se muestra como pasar información fuera de la transacción.



````
doc = "EmpleadoF"

transaction = db.transaction()
ref_emp = db.collection('Empleados').document(f"{doc}")

@firestore.transactional
def update_in_transaction(transaction, ref_emp):
    snapshot = ref_emp.get(transaction=transaction)
    nueva_edad = snapshot.get('EDAD')+1
    if nueva_edad < 45:
        transaction.update(ref_emp, {
          'EDAD': nueva_edad
        })
        return True
    else:
        return False


resultado = update_in_transaction(transaction, ref_emp)
if resultado:
    print('Edad actualizada')
else:
    print('Fuera del rango de edad.')
````

###**Escritura por lotes**

[Ver link](https://firebase.google.com/docs/firestore/manage-data/transactions)

Si no es necesario leer documentos en un conjunto de operaciones, se puede ejecutar varias operaciones de escritura como un solo bloque o lote. En este caso se combinan las siguientes operacioens:

- `set()`
- `update()`
- `delete()`

Vamos un ejemplo

````
batch = db.batch()

coll_name = 'Empleados'
doc_name1 = "EmpleadoL"
# Set EmpleadoL
ref_emp = db.collection(f'{coll_name}').document(f'{doc_name1}')
batch.set(ref_emp, {'Movil':3214456789})

# Update Edad de EmpleadoH
doc_name2 = "EmpleadoH"
ref_emp = db.collection(f'{coll_name}').document(f'{doc_name2}')
batch.update(ref_emp, {'EDAD': 26})

# Delete EmpleadoII
doc_name3 = "EmpleadoII"
ref_emp = db.collection(f'{coll_name}').document(f'{doc_name3}')
batch.delete(ref_emp)

# Commit the batch
batch.commit()
````
